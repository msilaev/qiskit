### Consider the Deusch problem with q Qubit $f(x) \rightarrow 0 \textrm{ or } 1 \textrm{ , where } x \textrm{ is } 0 \textrm{ or } 1$. Define 4 oracles :

Oracle 1 (constant):  $f_0 (0) = f_0 (1) =0$

```python
oracle1=qk.QuantumCircuit(2, name='oracle1')
oracle1.draw()
```

Oracle 2 (balanced):  $f_1 (0) =0$, $ f_1 (1) =1$

```python
oracle2=qk.QuantumCircuit(2, name='oracle2')
oracle2.cx(0,1)
oracle2.draw()
```

Oracle 3 (balanced):  $f_2 (0) =1$, $ f_2 (1) =0$

```python
oracle3=qk.QuantumCircuit(2, name='oracle3')
oracle3.cx(0,1)
oracle3.x(1)
oracle3.draw()
```

Oracle 4 (constant):  $f_3 (0) =1$, $ f_3 (1) =1$

```python
oracle4=qk.QuantumCircuit(2, name='oracle4')
oracle4.x(1)
oracle4.draw()
```

```python
circ= qk.QuantumCircuit(2,2)
#circ.h(0)
#circ.h([0,1])
circ.append(oracle2,[0,1])
backend = Aer.get_backend('statevector_simulator')
out = qk.execute(circ,backend).result().get_statevector()
plot_bloch_multivector(out)
```
